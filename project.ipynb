```python
def site_data_process_climate(site):
    file_path = f'./data/air_quality/PRSA_Data_{site}_20130301-20170228.csv'
    data = pd.read_csv(file_path)
    data.ffill(inplace=True)

    data = data[data['year']==2014]

    stats = data.groupby('month', as_index=False)[['TEMP', 'RAIN']].agg({'TEMP': [np.mean], 'RAIN': [np.sum]})
    stats.columns = ['month', 'TEMP', 'RAIN']
    return stats
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 假设 'data_aotizhongxin' 已经加载并且 'TEMP' 和 'time' 数据存在
data_Aotizhongxin['time'] = data_Aotizhongxin['year'].astype(str) + '-' + data_Aotizhongxin['month'].astype(str).str.zfill(2) + '-' + data_Aotizhongxin['day'].astype(str).str.zfill(2) + ' ' + data_Aotizhongxin['hour'].astype(str).str.zfill(2) + ':00:00'
temperature = data_Aotizhongxin['TEMP']  # 替换为实际的温度列名
time = data_Aotizhongxin['PRES']  # 假设时间列名为'Time'

# 取样数据以减小内存占用，取前1000个数据点（你可以根据需要调整）
temperature_sample = temperature.head(100)
time_sample = time.head(100)

# 生成适合绘制3D图的网格，使用采样后的数据
T, Time = np.meshgrid(temperature_sample, time_sample)  # 温度和时间网格

# 确保Pollution也是二维数组，与Time和T的形状相匹配
Pollution = data_Aotizhongxin['PM2.5'].head(10000).values.reshape(len(time_sample), len(temperature_sample))

# 创建3D图形
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 绘制3D曲面
surf = ax.plot_surface(Time, T, Pollution, cmap='viridis', edgecolor='none')

# 添加颜色条
fig.colorbar(surf)

# 设置轴标签
ax.set_xlabel('压力 (hPa)')
ax.set_ylabel('温度 (°C)')
ax.set_zlabel('污染物浓度 (μg/m^3)')

# 添加标题
ax.set_title('温度、压力和污染物浓度的关系')
plt.savefig('temp.png')

# 显示图形
plt.show()
```

```python
def map_month_to_season(month):
    seasons = {
        1: 'Winter',  # 1月为冬季
        2: 'Winter',  # 2月为冬季
        3: 'Spring',  # 3月为春季
        4: 'Spring',  # 4月为春季
        5: 'Spring',  # 5月为春季
        6: 'Summer',  # 6月为夏季
        7: 'Summer',  # 7月为夏季
        8: 'Summer',  # 8月为夏季
        9: 'Autumn',  # 9月为秋季
        10: 'Autumn',  # 10月为秋季
        11: 'Autumn',  # 11月为秋季
        12: 'Winter'   # 12月为冬季
    }
    return seasons.get(month, 'Invalid month')

data_Aotizhongxin['season'] = data_Aotizhongxin['month'].apply(map_month_to_season)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# 假设你的数据集是一个CSV文件，包含12个站点和6种污染物的浓度
# 文件名为 'beijing_air_quality.csv'
# 列名分别为 'Site1', 'Site2', ..., 'Site12', 'Pollutant1', 'Pollutant2', ..., 'Pollutant6'

# 数据标准化
scaler = StandardScaler()
pollutants_scaled = scaler.fit_transform(data)

# 使用K-Means算法进行聚类
# 假设我们想要将数据聚成3类
kmeans = KMeans(n_clusters=3, random_state=0).fit(pollutants_scaled)

# 将聚类结果添加到原始数据框中
data['Cluster'] = kmeans.labels_

# 可视化聚类结果（以两种污染物为例）
plt.figure(figsize=(10, 6))
plt.scatter(pollutants_scaled[:, 0], pollutants_scaled[:, 1], c=kmeans.labels_, cmap='viridis')
plt.xlabel('Pollutant1')
plt.ylabel('Pollutant2')
plt.title('K-Means Clustering of Air Quality Data')
plt.colorbar(label='Cluster')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import STL
# STL分解
data = data_Aotizhongxin['PM2.5']
stl = STL(data, seasonal=13, trend='spline', robust=True)
result = stl.fit()

# 提取分解结果
trend = result.trend
seasonal = result.seasonal
residual = result.resid

# 绘制原始数据
plt.figure(figsize=(14, 8))
plt.subplot(411)
plt.plot(data, label='Original', color='blue')
plt.legend(loc='best')
plt.title('Original Data')

# 绘制趋势
plt.subplot(412)
plt.plot(trend, label='Trend', color='green')
plt.legend(loc='best')
plt.title('Trend')

# 绘制季节性
plt.subplot(413)
plt.plot(seasonal, label='Seasonal', color='red')
plt.legend(loc='best')
plt.title('Seasonal')

# 绘制残差
plt.subplot(414)
plt.plot(residual, label='Residual', color='purple')
plt.legend(loc='best')
plt.title('Residual')

plt.tight_layout()
plt.show()
```

```python
def site_data_process(site):
    file_path = f'./data/air_quality/PRSA_Data_{site}_20130301-20170228.csv'
    data = pd.read_csv(file_path)
    data = data.ffill()
    data = data.query('year == 2013 & month == 3 & day == 7')
    return data['PM2.5'].mean()

site_list = ['Tiantan', 'Aotizhongxin', 'Wanliu', 'Changping', 'Dingling', 'Dongsi', 'Guanyuan', 'Gucheng', 'Huairou', 'Nongzhanguan', 'Shunyi', 'Wanshouxigong']
data={}
PM=[]
for site in site_list:
    mean_pm25 = site_data_process(site)
    PM.append(mean_pm25)  # 将站点和其PM2.5平均值添加到字典中
data['station'] = site_list
data['PM2.5'] = PM
```
